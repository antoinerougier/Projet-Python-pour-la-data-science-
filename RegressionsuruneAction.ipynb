```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://minio.lab.sspcloud.fr/arougier/test_autre.csv", sep=";")
df
```

```python
#on prend seulement la première colonne
#new_df = df.drop(df.columns[0, 2, 3, 4, 5], axis=1, inplace=True)
new_df=df[['open']]
new_df

#on prend les 742 premières valeurs
new_df2=new_df[0:742]
new_df2
```

```python
# On a donc pris à la main les 742 valeurs de la première action faisont maintenant le machine learning

X = new_df2[0:741]
Y = new_df2[1:741]

dataset = new_df2
X = dataset.iloc[0:741,0]
Y = dataset.iloc[1:742,0]


X_train = X.iloc[0:500].values
Y_train = Y.iloc[0:500].values



X_test = X.iloc[500:742].values
Y_test = Y.iloc[500:742].values

X_test
```

```python
#importation de la régression linéaire

from sklearn.linear_model import LinearRegression

reg = LinearRegression()

#on utilise fit pour entraîner le modèle


reg.fit(np.transpose([X_train]), np.transpose([Y_train]))

y_pred = reg.predict(np.transpose([X_test]))

#Graphique des prédictions

import matplotlib.pyplot as plt

plt.plot(y_pred, color='green')
plt.plot(Y_test, color='gray')
```
